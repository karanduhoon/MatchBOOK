```python
import kagglehub
from matplotlib.pyplot import plot_date
from seaborn import heatmap

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=True,ax=ax)

plt.xlabel("columns")
plt.ylabel("missing values")

plt.show()
```

```python
import numpy as np
```

```python
books['missing_desc'] = np.where(books['description'].isna(),1,0)
```

```python
books['age_of_the_book'] = 2025-books['published_year']
```

```python
col_of_interest = ['num_pages','age_of_the_book','missing_desc','average_rating']
```

```python
# because we are dealing with non-continous values spearman is more efficenient

corr_matrix = books[col_of_interest].corr(method="spearman")
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,10))
heat_map = sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
heat_map.set_title("correlation heatmap")
plt.show
```

```python
books[books['description'].isna() | books['num_pages'].isna()|books['average_rating'].isna()| books['published_year'].isna()]

```

```python
# lets remove the rows with missing values since it is not going to affect us much

book_clean = books[~books['description'].isna() &
 ~books['num_pages'].isna()&
 ~books['average_rating'].isna()&
 ~books['published_year'].isna()]
```

```python
book_clean
```

```python
book_clean['categories'].value_counts().reset_index().sort_values(by='count', ascending=False)
```

```python
book_clean['words_in_desc'] = book_clean['description'].str.split().str.len()
```

```python
book_clean
```

```python
book_clean.loc[book_clean['words_in_desc'].between(1,4), 'description']
```

```python
book_clean.loc[book_clean['words_in_desc'].between(15,35), 'description'].count()
```

```python
book_clean_25words = book_clean[book_clean['words_in_desc']>=23]
```

```python
book_clean_25words['subtitle'].isnull().sum()
```

```python
book_clean_25words["title_and_subtitle"] = (
    np.where(book_clean_25words["subtitle"].isna(), book_clean_25words["title"],
             book_clean_25words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
book_clean_25words
```

```python
book_clean_25words["tagged_desc"] = book_clean_25words[["isbn13",'description']].astype(str).agg(" ".join, axis=1)
```

```python
book_clean_25words
```

```python
book_clean_25words.drop(['subtitle','missing_desc','words_in_desc'], axis=1)
```

```python
book_clean_25words.to_csv('books_cleaned.csv', index=False)
```
